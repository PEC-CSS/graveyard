```python
import numpy as np
import pandas as pd
```

```python
columns = ["Population (Millions)", "Size (Km^2)"]
index = ["Codeville", "Binaryburg", "Bugland"]
```

```python
data = np.array([[50, 100000], [100, 220000], [150, 350000]])
data
```

```python
df = pd.DataFrame(data=data, columns=columns, index=index, dtype=int)
df
```

```python
df["Population Density (Millions per Km^2)"] = df["Population (Millions)"] / df["Size (Km^2)"]
df
```

```python
df.index.name = 'City'
df
```
