```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([10,20,30,40,50,60,77])
y = np.array([100,20,1,19,22,22,22])
```

```python
plt.scatter(x, y, color='red', marker='x')
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
plt.plot(x, y, color="green", linestyle="dashed", linewidth=2, marker="o")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
plt.step(x, y, color='blue', linewidth=2, marker="o", linestyle="solid", where="post")
for i in range(len(x)):
  plt.vlines(x[i], 0, y[i], linestyle="dashed", color="grey")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
plt.stem(x,y)
```
